# Ngrams

Given the name of a study carrel in the Distant Reader Library as well as a query term, perform ngram analysis against the carrel.

Approximately 3,000 study carrels have been previously created, saved, and made available on a website called the Distant Reader Library. It is possible to apply rudimentary ngram analysis against these carrels for the purposes of addressing real research questions or evaluating whether or not the remote carrel is of continued interest. This notebook demonstrates how.


```python
# configure
CARREL = 'homer'
QUERY  = 'jove'
```

```python
# require
import rdr
```

```python
# get, tabulate, and output all unigrams in the remote carrel
print( rdr.ngrams( CARREL, size=1, count=True, location='remote' ) )
```

```python
# get, tabulate, and output all bigrams in the remote carrel
print( rdr.ngrams( CARREL, size=2, count=True, location='remote' ) )
```

```python
# get and output a word cloud of all bigrams in the remote carrel
rdr.ngrams( CARREL, size=2, count=True, location='remote', wordcloud=True )
```

```python
# get and output a word cloud of all bigrams containing the configured query
rdr.ngrams( CARREL, size=2, count=True, location='remote', wordcloud=True, query=QUERY )
```
