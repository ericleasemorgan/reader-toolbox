# Email addresses

Given a study carrel, output information about email addresses.

Email addresses are extracted from carrel content. Knowing what email addresses exist helps address the question, "Who are you gonna call?" This notebook demonstrates this concept.

Two additional things. First, study carrel named homer contains zero email addresses. Therefore a different carrel is used. More specifically, a carrel named 'ital-2006-2010', which includes about 160 scholarly journal articles on the topic of computers and libraries. The articles include many email address. Second, please use the output of these functions responsibly.

```python
# configure
CARREL = 'ital-2006-2010'
FILTER = 'oclc'
```

```python
# require
import rdr
```

```python
# check to see if the carrel exists; if not, then download it
try    : rdr.checkForCarrel( CARREL )
except : rdr.download( CARREL )
```

```python
# count, tabulate, and output all email addresses
result = rdr.addresses( CARREL, count=True )
print( result )
```

```python
# count, tabulate, filter, and output email addresses
result = rdr.addresses( CARREL, count=True, like=FILTER )
print( result )
```
