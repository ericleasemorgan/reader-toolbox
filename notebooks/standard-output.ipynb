# Standard Output

This notebook outputs the generic report generated by the study carrel creation process. If it exists, then the report includes rudimentary summary information regarding the carrel's content.

```python
# configure
STANDARDOUTPUT = 'standard-output.txt'
```

```python
# require
from pathlib import Path
```

```python
# initialize
carrel         = Path().absolute().parent
standardoutput = carrel/STANDARDOUTPUT
```

```python
# open the file, output the content, and done
with open( standardoutput ) as handle : print( handle.read() )
exit()
```
