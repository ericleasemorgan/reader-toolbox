# Parts-Of-Speech

Given the name of a study carrel, output information based on parts-of-speech.

```python
# configure
CARREL = 'homer'
```

```python
# require
import rdr
```

```python
# count, tabulate, and output the different parts-of-speech (nouns, verbs, adjectives, etc.)
print( rdr.pos( CARREL, count=True ) )
```

```python
# count, tabulate, and output just the nouns
print( rdr.pos( CARREL, select='words', like='NOUN', count=True ) )
```

```python
# count and tabulate lematized nouns
print( rdr.pos( CARREL, select='lemmas', like='NOUN', count=True ) )
```

```python
# count and tabulate the lematized nouns and output the result as a wordcloud
rdr.pos( CARREL, select='lemmas', like='NOUN', count=True, wordcloud=True )
```
