# Readability

Given a study carrel, output information regarding the carrel's readability.

The "readability" of a text is computed using things like the length of words, length of sentences, number of words, size of vocabulary, etc. The Toolbox implements a readability measure called Flesch, and a score of 0 means nobody can read the document, and a score of 100 means everybody can read the document. In general, Dr. Suess stories have a high readaiblity score and insurance docments have a low score. Shakespear's Sonnets have a high readabilitiy score (about 90) because the words and number of words are small. Many novels have readability score in the 70's. Scholarly journal articles are a bit lower than that. If the readability score is very low, say below 50, then that is usually an indication that the underlying optical character recognition (OCR) of your documents is poor.

This notebook demonstrates who to determine the readability of the given carrel.

```python
# configure
CARREL = 'homer'
```

```python
# require
import rdr
```

```python
# get and output the average readabilitiy score of the carrel as a whole
flesch = rdr.extents( CARREL, 'flesch' )
print( '  readability:', str( flesch ) )
```

```python
# output a list of the carrel's documents and their readability score
print( rdr.flesch( CARREL ) )
```

```python
# get the readability of all documents, and output the results as boxplot
result = rdr.flesch( CARREL, output='boxplot') 
```

```python
# get the readability of all documents, and output the results as histogram
result = rdr.flesch( CARREL, output='histogram') 
```
