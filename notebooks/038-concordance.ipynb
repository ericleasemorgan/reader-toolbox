# Concordance

Given the name of a study carrel and a query, apply simple keyword-in-context (KWIC) searching to the carrel.

Concordances are one of the oldest of text mining techniques. Developed in the 13th Century, they were used to observe how a given word or phrase were used in conjunction with other words. The Toolbox implements such functionality, it is analgous to the modern-day ^F functionality found in many word procesors, and this notebook demonstrates how.


```python
# configure
CARREL       = 'homer'
QUERY        = 'ulysses'
ANOTHERQUERY = 'ulysses was'
```

```python
# require
import rdr
```

```python
# do the work
lines = rdr.concordance( CARREL, query=QUERY )
for line in lines : print( line )
    
```

```python
# do the work again
lines = rdr.concordance( CARREL, query=ANOTHERQUERY )
for line in lines : print( line )
```
