# Keywords

Given the name of a carrel, output information regarding statistically significant keywords.

The venerable TF/IDF (Term-Frequency/Inverse Document Frequency) algorithm is useful for many things, and one of them is denoting the aboutness of a document. The Toolbox employs a variation of TF/IDF called Yake to perform the this function. Thus, statistically significant words have been identified for each of the documents in a study carrrel, and you can use these words to learn about individual documents as well as the carrel as whole. This notebook introduces how this can be done.

```python
# configure
CARREL = 'homer'
```

```python
# require
import rdr
```

```python
# get and output a list of all the keywords in given carrrel
keywords = rdr.keywords( CARREL )
print( keywords )
```

```python
# get, count, tabulate, and output all the keywords in the given carrel
print( rdr.keywords( CARREL, count=True ) )
```

```python
# get, count, tabulate, and output all the keywords in the given carrel as a wordcloud
result = rdr.keywords( CARREL, count=True, wordcloud=True )
```
