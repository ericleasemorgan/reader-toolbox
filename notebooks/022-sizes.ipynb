# Sizes

Given a study carrel, measure and visualize the number of items as well as the number words in a carrel.

It is very important to know the size of your carrel, measured both in the number of words as well as the number of items (documents). This is true because different modeling techniques can be applied to smaller carrels, and other modeling techniques can only be applied to larger carrels. This notebook demostrates how you can learn about the size of yourl carrel.

When doing this sort of analysis, it is important to be able to compare the results to known items. For example, combined, Homer's Iliad and Odyssey are about 270,000 (.27 million) words long. The Bible is about .8 million words long. Now, ask yourself, "How big are Homer's works compared to the Bible?" The typical scholarly journal article is somewhere between 4,000 and 8,000 (.004 and .008 million) words long. Most people who do natural langauge processing do not consider their corpus to be large until it reaches at least 1 million words, and even that is small.


```python
# configure
CARREL = 'homer'
```

```python
# require
import rdr
```

```python
# get the number of items and the number of words in the given carrel
items = rdr.extents( CARREL, 'items' )
words = rdr.extents( CARREL, 'words' )

# output
print( '  number of items:', str( items ) )
print( '  number of words:', str( words ) )
```

```python
# output a list of the carrel's documents and their sizes measured in words
print( rdr.sizes( CARREL ) )
```

```python
# output a boxplot illustrating the relative sizes of the documents
result = rdr.sizes( CARREL, output='boxplot' )
```

```python
# output a histogram illustrating the relative sizes of the documents
result = rdr.sizes( CARREL, output='histogram' )
```
