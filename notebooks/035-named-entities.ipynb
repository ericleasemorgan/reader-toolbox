# Named Entities

Given the name of a carrel, output information regarding the carrel's named entitites.

Named entitites are deemed as real-world things. The most common are people, organizations, places, and locations. Other entities include dates, times, money amounts, different types of numbers, etc. Knowing what named entitites exist in a carrel helps you address many questions, such as who, where, when, and how many. This notebook demonstrates how to identify, count, and tabulate these things.


```python
# configure
CARREL = 'homer'
```

```python
# require
import rdr
```

```python
# count, tabulate, and output the frequency of the different types of entities
print( rdr.entities( CARREL, count=True ) )
```

```python
# count, tabulate, and output the frequency of PERSONs
print( rdr.entities( CARREL, select='entity', like='PERSON', count=True ) )
```

```python
# count, tabulate PERSONS, and output as a word cloud
rdr.entities( CARREL, select='entity', like='PERSON', count=True, wordcloud=True )
```

```python
# address the question, "What places are mentioned?", and output the result as a wordcloud
rdr.entities( CARREL, select='entity', like='GPE', count=True, wordcloud=True )
```
