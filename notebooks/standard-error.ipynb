# Standard Error

This notebook outputs the error log generated by the study carrel generation process. If something went wrong with the creation of the carrerl, then clues to the failure ought to be listed here. Be forewarned. The output can be quite long.


```python
# configure
STANDARDERROR = 'standard-error.txt'
```

```python
# require
from pathlib import Path
```

```python
# initialize
carrel = Path().absolute().parent
standarderror = carrel/STANDARDERROR
```

```python
# open the file, output the content, and done
with open( standarderror ) as handle : print( handle.read() )
exit()
```
